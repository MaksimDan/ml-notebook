# Introduction to TensorFlow

TensorFlow is a powerful open source software library for numerical computation, particularly well suited and fine-tuned for large-scale Machine Learning. It is based on graphs in python where nodes in the graph are operations, and leaf nodes are variables. This is over simplified, but to give a basic idea, we can computations such as this:


```python
from graphviz import Digraph


g = Digraph('G')
g.attr('node', shape='circle')
g.node_attr.update(color='lightblue2', style='filled')

g.node('+0', label='+')
g.node('x0', label='x')
g.node('+1', label='+')
g.node('x1', label='x')

# g.attr('node', shape='square')
g.node('X', label='X')
g.node('y', label='y')
g.node('2', label='2')

g.edge('x0', '+0')
g.edge('+1', '+0')
g.edge('x1', 'x0')

g.edge('X', 'x1')
g.edge('X', 'x1')
g.edge('y', 'x0')
g.edge('y', '+1')
g.edge('2', '+1')

g
```

Which as an abstract syntax tree reads as $f(x,y)=x^2y + y + 2$. By running a post order traversal with a map that keys variable names to their value, we can solve any kind of expression. 

One of the benefits is that that expressions can be broken up into independent chunks and executed in parallel.
