# Application - Data Generation

We can use a guassian model to sample new instances, and make predictions on top of those.

```python
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn.datasets import make_moons
from sklearn.mixture import GaussianMixture
from mltoolbox.draw.color import map_labels_to_random_color


X, y = make_moons(100, noise=0.05)
plt.scatter(X[:, 0], X[:, 1], color=map_labels_to_random_color(y))
plt.title('make_moons(100, noise=0.05)')
```

```python
gm = GaussianMixture(n_components=2, n_init=10)
gm.fit(X)
X_imputed, y_imputed = gm.sample(200)
plt.scatter(X[:, 0], X[:, 1])
plt.scatter(X_imputed[:, 0], X_imputed[:, 1], color=map_labels_to_random_color(y_imputed))
plt.title('200 Sampled GaussianMixture(n_components=2, n_init=10) overlay on Original Dataset')
```
